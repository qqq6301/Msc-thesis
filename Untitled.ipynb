```python
#load numpy and pandas 
import pandas as pd
from pandas import DataFrame
import numpy as np 
import datetime as dt
#load library for modelling 
import statsmodels 
from statsmodels.tsa.arima.model import ARIMA
#add plot
from matplotlib import pyplot
```

```python
#Read in the data, data is download from NHS website, and is pre-processed to combine most recent 5-year data in 
#a single file 
```

```python
df = pd.read_excel ('5-year-data-combined.xlsx')
```

```python
#Data Understanding - how many hosptials are there in the dataset?
```

```python
#Data Understanding - what are the value range of "STANRARD" , what is respective meaning of each possible value?
```

```python
#Subset the data, to foucus on a hosptial, one cancer type, and one referral type.
```

```python
df_subset1= df[(df["STANDARD"]=="2WW") & (df["CANCER TYPE"]=="Suspected lower gastrointestinal cancer")]
df_subset1_RXN=df_subset1[df_subset1["ORG CODE"]=="RXN"]
```

```python
#Handle data quality issue, one data quality issue if found. - There could be more entries for one same time period. 
#Need to handle, otherwise not able to make correct plot to show trends. 
```

```python
df_subset1_RXN2=df_subset1_RXN.groupby(['PERIOD','YEAR','STANDARD','ORG CODE','CANCER TYPE']).sum()
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
```

```python
#Add a calculated column called percentage, which is a metric to measure how quickly a hosptial handles incoming cancer referrals. 
```

```python
df_subset1_RXN2["PERCENTAGE"]=df_subset1_RXN2["WITHIN STANDARD"]/df_subset1_RXN2["TOTAL TREATED"] 
```

```python
#Make time-series plot 
```

```python
#First plot use percentage as Y-axis, oberservation: (1)When the pandanmic hits UK hardest, spring 2020, the percentage metric is still good.
#However, we also need to draw another plot to see how many paitents are referred during this period. (2)The percentage metric is low in early
#2021, however, since it is Christimas holiday, it is understandable and not a surprise. 
```

```python
pyplot.plot(df_subset1_RXN2["PERIOD"], df_subset1_RXN2["PERCENTAGE"])
```

```python
#First plot use total referred patients as Y-axis,we can see the number of incomming referral patients significantly decrease in Spring,2020 
#due to pandanmic and lockdown measures. 
```

```python
pyplot.plot(df_subset1_RXN2["PERIOD"], df_subset1_RXN2["TOTAL TREATED"])
```
