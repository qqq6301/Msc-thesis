```python
#Subset the data, to foucus on a hosptial, one cancer type, and one referral type.
```

```python
df_subset1= df[(df["STANDARD"]=="2WW") & (df["CANCER TYPE"]=="Suspected lower gastrointestinal cancer")]
df_subset1_RXN=df_subset1[df_subset1["ORG CODE"]=="RXN"]
```

```python
#Handle data quality issue, one data quality issue if found. - There could be more entries for one same time period. 
#Need to handle, otherwise not able to make correct plot to show trends. 
```

```python
df_subset1_RXN2=df_subset1_RXN.groupby(['PERIOD','YEAR','STANDARD','ORG CODE','CANCER TYPE']).sum()
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
df_subset1_RXN2.reset_index(level=0, inplace=True)
```

```python
#Add a calculated column called percentage, which is a metric to measure how quickly a hosptial handles incoming cancer referrals. 
```

```python
df_subset1_RXN2["PERCENTAGE"]=df_subset1_RXN2["WITHIN STANDARD"]/df_subset1_RXN2["TOTAL TREATED"] 
```

```python
df_subset1_RXN_ts = df_subset1_RXN.loc[:,['PERIOD', 'TOTAL TREATED']]
```

```python
training_data=df_subset1_RXN_ts #Omit train-test split for week 1  
```

```python
training_data_series = pd.Series(training_data["TOTAL TREATED"])
training_data_series.index=training_data["PERIOD"]
```

```python
# fit model
model = ARIMA(training_data_series,order=(5,1,0))
model_fit = model.fit()
```

```python
# summary of fit model
print(model_fit.summary())
```

```python
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
```
